```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
# Load the dataset from a local CSV file
#iris = pd.read_csv(r'C:\Users\91808\Desktop\Data\Iris.csv')
# Load the data set directly from the link
iris ='https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv'
iris = pd.read_csv(iris)
```

```python
#Display the columns in the dataset.
for column in iris.columns:
    print(column)
```

```python
# Calculate the mean of each column of the iris dataset
mean = iris.select_dtypes(include='number').mean()
print(mean)
```

```python
# Check for null values if any
null_values = iris.isnull().sum()
print(null_values)
```

```python
# Calculate the average sepal length for each species
avg_sepal_length = iris.groupby('species')['sepal_length'].mean().reset_index()

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(avg_sepal_length['species'], avg_sepal_length['sepal_length'], marker='o', linestyle='-')
plt.title('Average Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.grid(True)
plt.show()
```

```python
# Calculate the average petal length for each species
avg_petal_length = iris.groupby('species')['petal_length'].mean().reset_index()

# Plot the bar chart
plt.figure(figsize=(10, 6))
plt.bar(avg_petal_length['species'], avg_petal_length['petal_length'], color=['red', 'green', 'blue'])
plt.title('Average Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length')
plt.show()
```

```python
# Define colors for each species
species_colors = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}
colors = iris['species'].map(species_colors)

# Plot the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(iris['sepal_length'], iris['petal_length'], c=colors)
plt.title('Sepal Length vs. Petal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()
```
