```python
import pandas as pd
```

```python
# Load the dataset
file_path = '/content/Students_Performance.csv'
df = pd.read_csv(file_path)
```

```python
# Display the first few rows
df.head()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Display basic statistics
df.describe(include='all')
```

a. Find out how many males and females participated in the test

```python
# Count the number of males and females
gender_counts = df['gender'].value_counts()
gender_counts
```

b. What do you think about the students' parental level of education?

```python
# Analyze the distribution of parental levels of education
parental_education_counts = df['parental level of education'].value_counts()
parental_education_counts
```

c. Who scores the most on average for math, reading and writing based on i. Gender ii. Test preparation course

```python
# Calculate average scores based on gender
average_scores_gender = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
average_scores_gender

# Calculate average scores based on test preparation course
average_scores_preparation = df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean()
average_scores_preparation
```

 d. What do you think about the scoring variation for math, reading and writing based on i. Gender ii. Test preparation course

```python
# Assess the variation in scores based on gender
variation_scores_gender = df.groupby('gender')[['math score', 'reading score', 'writing score']].std()
variation_scores_gender

# Assess the variation in scores based on test preparation course
variation_scores_preparation = df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].std()
variation_scores_preparation
```

 e. The management needs your help to give bonus points to the top 25% of students based on their math score, so how will you help the management to achieve this

```python
# Determine the cutoff for the top 25% in math scores
math_score_cutoff = df['math score'].quantile(0.75)

# Identify students in the top 25%
top_25_percent_students = df[df['math score'] >= math_score_cutoff]
top_25_percent_students
```

```python
# Add bonus points (e.g., 10 points) to the top 25% students' math scores
df['math score'] = df.apply(lambda row: row['math score'] + 10 if row['math score'] >= math_score_cutoff else row['math score'], axis=1)
df
```

```python
import pandas as pd
```

Step 1: Load and Explore the Dataset

```python
# Load the dataset
sales_data = pd.read_csv('/content/Sales_add.csv')
```

```python
# Display the first few rows
sales_data.head()

# Check for missing values
sales_data.isnull().sum()

# Display basic statistics
sales_data.describe(include='all')
```

a. The company wishes to clarify whether there is any increase in sales after stepping into digital marketing

Step 2: Test Hypothesis for Increase in Sales
Formulate Hypotheses
Null Hypothesis (H0): There is no increase in sales after digital marketing.
Alternative Hypothesis (H1): There is an increase in sales after digital marketing.
Perform Paired Sample T-test

```python
from scipy.stats import ttest_rel

# Assuming the dataset has columns 'Sales_Before' and 'Sales_After'
sales_before = sales_data['Sales_before_digital_add(in $)']
sales_after = sales_data['Sales_After_digital_add(in $)']
#sales_before = df['Sales_before_digital_add']
#sales_after = df['Sales_After_digital_add']

# Perform paired sample t-test
t_stat, p_value = ttest_rel(sales_before, sales_after)

# Display the results
t_stat, p_value
```

Step 3: Test Dependency between Region and Manager
Formulate Hypotheses
Null Hypothesis (H0): Region and Manager are independent.
Alternative Hypothesis (H1): Region and Manager are dependent.
Perform Chi-square Test of Independence

```python
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(sales_data['Region'], sales_data['Manager'])

# Perform Chi-square test of independence
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Display the results
chi2_stat, p_value
```
