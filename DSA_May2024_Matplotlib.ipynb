```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40]
y=[20,30,40,50]

plt.title("Simple Line Plot")
plt.plot(x,y,marker='s',color='yellow',linewidth=3,linestyle='solid')
plt.xlabel("X-axis data")
plt.ylabel("y-axis data")
```

```python
students = ["Akash","maya","Jane","Anna","Sooraj","Rahim"]
marks = [89,45,78,67,55,71]

plt.xlabel("students")
plt.ylabel("Marks")
plt.title("Class Records")
plt.xticks(rotation=45)
plt.plot(students,marks,linestyle="dotted",marker="*")
```
