```python
#one hot encoding
categorical_columns=['sex','embarked','class','who','embark_town','alone']
one_hot_enc=OneHotEncoder(sparse=False,drop='first')
encoded_data=one_hot_enc.fit_transform(data[categorical_columns])
encoded_data
```

```python
encoded_categorical_df=pd.DataFrame(encoded_data,columns=one_hot_enc.get_feature_names_out(categorical_columns))
encoded_categorical_df
```

```python
data_encoded=pd.concat([data.drop(columns=categorical_columns),encoded_categorical_df],axis=1)
data_encoded.head()
```
